# EXPLORATORY DATA ANALYSIS

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('SampleSuperstore.csv')
df
```

```python
df=df.drop_duplicates()
df
```

```python
df.drop('Country', axis=1, inplace=True)
df.tail(1)
```

```python
df.info()
```

```python
df.describe()
```

```python
corr=df.corr()
corr
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(corr,annot=True)
plt.show()
```

```python
df['Sub-Category'].value_counts()
```

DISTRIBUTION PLOTS

```python
num_bins = 8
plt.hist(df['Ship Mode'], num_bins, facecolor ='forestgreen' , alpha = 0.8)
plt.show()
```

```python
num_bins = 8
plt.hist(df['Category'], num_bins, facecolor ='springgreen' , alpha = 0.8)
plt.show()
```

```python
num_bins = 15
plt.hist(df['Discount'], num_bins, facecolor ='seagreen' , alpha = 0.8)
plt.show()
```

DATA VISUALIZTION


```python
plt.figure(figsize=(16,8))
plt.bar('Sub-Category','Category',data=df)
plt.xticks(rotation=90)
```

```python
df1 = df['State'].value_counts()
df1.plot(kind='bar',figsize=(15,5))
plt.ylabel('Frequency / Number of deals')
plt.xlabel('States')
plt.title('State Wise Dealings', fontsize = 12)
plt.show()
```

```python
print(df['State'].value_counts())
```

```python
storegroup=df.groupby("Segment")["Quantity"].sum().reset_index()
storegroup
```

```python
plt.figure(figsize=(10,8))
df['Sub-Category'].value_counts().plot.pie(autopct='')
plt.show()
```

Analysis of Profits

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Sub-Category',y='Profit',data=df)
plt.xticks(rotation=90)
plt.show()
```
